## Homework Question 1

Save the summary statistics of all numerical columns and all categorical columns to only one Excel spreadsheet file with two worksheets, one for numerical and one for categorical columns.

```python
import pandas as pd

# Assuming df_2020 is your DataFrame

# Summary statistics for numerical columns
numerical_summary = df_2020.select_dtypes(include=["number"]).agg(['count', 'mean', 'std', 'min', 'max'])

# Adding quantiles (percentiles) separately
quantiles = df_2020.select_dtypes(include=["number"]).quantile([0.25, 0.5, 0.75])
quantiles.index = ['25%', '50%', '75%']

# Concatenating quantiles to numerical summary
numerical_summary = pd.concat([numerical_summary, quantiles])

# Summary statistics for categorical columns
categorical_summary = df_2020.select_dtypes(include=["object"]).describe()

# Save both summaries into an Excel file with two worksheets
with pd.ExcelWriter("summary_statistics_alt_fixed.xlsx", engine='openpyxl') as writer:
    numerical_summary.to_excel(writer, sheet_name="Numerical Stats")
    categorical_summary.to_excel(writer, sheet_name="Categorical Stats")

print("Summary statistics saved to Excel successfully!")
```

## Homework Question 3 (bonus, will cover next week)

step 1 - create a column "Life Expectancy Quartile" similar to the GDP per Capita Quartile above.
Step 2 - Create a column "health and wealth status" based on the following definition
- rich and healthy (1st GDP per capita quartile and 1st life expentancy quartile)
- poor and unhealthy ((4th GDP per capita quartile and 4th life expentancy quartile)
- rich and unhealthy (1st GDP per capita quartile and 4th life expentancy quartile)
- poor and healthy ((4th GDP per capita quartile and 1st life expentancy quartile)
- Other

```python
import pandas as pd

# Define the quartile boundaries
bins = [0, 63.026, 71.089, 76.907, float('inf')]
labels = ["1st Quartile (<=25%)", "2nd Quartile (25-50%)", "3rd Quartile (50-75%)", "4th Quartile (>75%)"]

# Use pd.cut to assign quartiles based on the Life Expectancy values
df_2020["Life Expectancy Quartile"] = pd.cut(df_2020["Life Expectancy"], bins=bins, labels=labels, right=True)

# Sample the result
df_2020.sample(5)
```

```python
def health_wealth_status(row):
    status_mapping = {
        ("1st Quartile (<=25%)", "1st Quartile (<=25%)"): "rich and healthy",
        ("4th Quartile (>75%)", "4th Quartile (>75%)"): "poor and unhealthy",
        ("1st Quartile (<=25%)", "4th Quartile (>75%)"): "rich and unhealthy",
        ("4th Quartile (>75%)", "1st Quartile (<=25%)"): "poor and healthy",
    }

    gdp_quartile = row["GDP Quartile"]
    life_expectancy_quartile = row["Life Expectancy Quartile"]

    return status_mapping.get((gdp_quartile, life_expectancy_quartile), "Other")

df_2020["Health and Wealth Status"] = df_2020.apply(health_wealth_status, axis=1)

df_2020.sample(5)
```
