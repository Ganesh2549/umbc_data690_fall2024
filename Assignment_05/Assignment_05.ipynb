Pick a few countries of your interest (for example, G7, BRICS, Nordic countries).
- 1 Create a line chart showing the trend of Life Expectancy over time for comparing those countries.
- 2 Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.
- 3 Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

```python
#Create a line chart showing the trend of Life Expectancy over time for comparing those countries.
```

```python
# Define the list of G7 nations
g7_nations = ["Canada", "France", "Germany", "Italy", "Japan", "United Kingdom", "United States"]

# Select only rows where the country is in the G7 list
df_g7_nations = df[df["Country"].isin(g7_nations)]

# Remove rows with missing data in life expectancy and population columns
df_g7_nations_cleaned = df_g7_nations.dropna(subset=["Life expectancy at birth, total (years)", "Population, total"])
```

```python
# Plot a line graph showing life expectancy trends for G7 nations over time
fig_life_expectancy = px.line(
    df_g7_nations_cleaned,
    x="Year",
    y="Life expectancy at birth, total (years)",
    color="Country",
    title="Life Expectancy Trends in G7 Nations",
    labels={"Life expectancy at birth, total (years)": "Life Expectancy (Years)"}
)

# Display the chart
fig_life_expectancy.show()
```

```python
#2 Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.
```

```python
# Set the target year for analysis
target_year = 2019

# Filter the DataFrame for the specified year
df_g7_target_year = df_g7_nations[df_g7_nations["Year"] == target_year]

# Remove entries with missing values for GDP per capita or life expectancy
df_g7_target_year_cleaned = df_g7_target_year.dropna(subset=["GDP per capita (current US$)", "Life expectancy at birth, total (years)"])

# Generate a bubble chart comparing GDP per capita to life expectancy
fig_bubble_chart = px.scatter(
    df_g7_target_year_cleaned,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    size="Population, total",
    color="Country",
    title=f"Bubble Chart: GDP per Capita vs Life Expectancy in {target_year}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

# Render the bubble chart
fig_bubble_chart.show()
```

```python
#3 Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions
```

```python
# Eliminate rows with missing data for GDP per capita, life expectancy, or total population
df_cleaned = df.dropna(subset=["GDP per capita (current US$)", "Life expectancy at birth, total (years)", "Population, total"])

# Calculate the average GDP per capita, average life expectancy, and total population for each region and year
df_grouped_by_region = df_cleaned.groupby(["Year", "Region"]).agg({
    "GDP per capita (current US$)": "mean",
    "Life expectancy at birth, total (years)": "mean",
    "Population, total": "sum"
}).reset_index()

# 3.1: Create a line graph showing life expectancy trends by region over time
fig_life_expectancy_trends = px.line(
    df_grouped_by_region,
    x="Year",
    y="Life expectancy at birth, total (years)",
    color="Region",
    title="Life Expectancy Trends Over Time by Region",
    labels={"Life expectancy at birth, total (years)": "Life Expectancy (Years)"}
)

# Display the life expectancy trend chart
fig_life_expectancy_trends.show()
```

```python
# 3.2: Generate a bubble chart for average GDP per capita and life expectancy for the most recent year in the dataset (2022)
latest_year = df_grouped_by_region["Year"].max()

# Filter the DataFrame for the most recent year
df_latest_year_data = df_grouped_by_region[df_grouped_by_region["Year"] == latest_year]

# Create a bubble chart visualizing the aggregated data
fig_bubble_chart_gdp_life_expectancy = px.scatter(
    df_latest_year_data,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    size="Population, total",
    color="Region",
    title=f"Bubble Chart: GDP per Capita vs Life Expectancy in {latest_year}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

# Render the bubble chart
fig_bubble_chart_gdp_life_expectancy.show()
```
