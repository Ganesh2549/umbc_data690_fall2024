Let's experiment with column "10. What do you expect to get out of this course?" to see if we can determine most popular leisure time actvities.

````python
list_answers = df["9. What do you enjoy doing in your leisure time?"].tolist()

list_answers
````

We will start with an empty dictionary like this:
```python
word_freq = {}
```

We hope to end up with something like this:
```python
word_freq{
    "movies" : 3,
    "reading": 10,
    "sports": 3
}
```

````python
word_freq = {}

for answer in list_answers:
    list_words = answer.split()
    for word in list_words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

word_freq
````

## Assignment 03 Part 2

Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

````python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

import string
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Extract the responses from the relevant column
career_goals = df["11. What is your long-term career goal?"].dropna().tolist()

# Use NLTK's English stopwords set
stop_words = set(stopwords.words('english'))

# Combine punctuation and stopwords
stop_words.update({
    "career", "goal", "long", "term", "drive", "using", "work", "want", "aim", "problems"
})

# Join all responses into one large string and tokenize it
all_text = " ".join(career_goals).lower()

# Tokenize and filter out stopwords + punctuation in one step
words = [word for word in word_tokenize(all_text) if word not in stop_words and word not in string.punctuation]

# Use Counter to find the most common words
word_freq = Counter(words)

# Get the 30 most common words
top_goals = word_freq.most_common(30)

# Display results
print("Top Long-term Career Goals:")
for word, count in top_goals:
    print(f"{word}: {count}")


````

word frequency highlights the most commonly mentioned terms, with "analyst" and "engineer" being frequent. This suggests that many respondents have career aspirations related to software engineering, such as becoming developers. Additionally, frequent mentions of "machine" and "learning" indicate an interest in roles centered on AI and machine learning. The backgrounds of students in data science programs further reflect a strong inclination toward technical careers, particularly in areas like data science, engineering, and software development. This suggests that a significant number of students are aiming for roles in these fields.
